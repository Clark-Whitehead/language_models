```python
import torch
import matplotlib.pyplot as plt
```

```python
names = open('../data/names.txt', 'r').read().splitlines()
names[:5]
```

```python
b = {}
for name in names:
    chs = ['.'] + list(name) + ['.']
    for bigram in zip(chs, chs[1:]):
        b[bigram] = b.get(bigram, 0) + 1
        
b
```

```python
sorted(b.items(), key = lambda x : -x[1])
```

```python
chars = sorted(list(set(''.join(names))))
stoi = {s:i+1 for i, s in enumerate(chars)}
stoi['.'] = 0
stoi
```

```python
b
```

```python
itos = {i:s for s, i in stoi.items()}
itos
```

```python
N = torch.zeros(27, 27, dtype=torch.int64)
```

```python
for i in range(27):
    for j in range(27):
        N[i, j] = b.get((itos[i], itos[j]), 0)
        
N
```

```python
plt.figure(figsize=(16, 16))
plt.imshow(N, cmap="Blues")

for i in range(27):
    for j in range(27):
        chrset = itos[j] + itos[i]
        plt.text(i, j, N[j, i].item(), ha="center", va="top")
        plt.text(i, j, chrset, ha="center", va="bottom")
```

```python
P = (N+1).float()
P /= P.sum(1, keepdim=True)
```

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(5):

    out = []
    idx = 0
    while True:
        p = P[idx]
        idx = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(itos[idx])
        if idx == 0:
            break

    print(''.join(out))
```

```python
count = 0
log_likelihood = 0.0
for name in ["clark"]:
    chs = ['.'] + list(name) + ['.']
    for bigram in zip(chs, chs[1:]):
        count += 1
        idx1 = stoi[bigram[0]]
        idx2 = stoi[bigram[1]]
        prob = P[idx1, idx2]
        logprob = torch.log(prob)
        log_likelihood += logprob
#         print(f'{bigram}: {prob:.4f} {logprob:.4f}')
        
print()
        
print(f'{-log_likelihood / count}')
```
