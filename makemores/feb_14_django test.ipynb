```python
data = open('../data/lana_questions.txt', 'r', encoding='utf-8').read()
data
```

```python
import re

# remove '--------------' lines
data = re.sub('-+', '', data)

data = re.sub('—', '', data)

# remove extra newlines
data = re.sub('\n{3,}', '\n\n', data)

# print the cleaned text
print(data)
```

```python
file = open('../data/lana_questions_clean.txt', 'w')
file.write(data)
file.close()
```
